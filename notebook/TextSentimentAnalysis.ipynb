Now we will remove contractions and special characters from the dataset. We will also convert all words to lowercase. This is done because words like 'won't' and 'will not' are the same, but will not be treated as such. The same can be said for words like 'Go' and 'go'. In order to get a true word count, we must clean the data.  We also seen that the word "br" occured frequently in our data. This word obviously holds no value. It will also be removed.

```r
# function to expand contractions in an English-language source
fix.contractions <- function(doc) {
  # "won't" is a special case as it does not expand to "wo not"
  doc <- gsub("won't", "will not", doc)
  doc <- gsub("can't", "can not", doc)
  doc <- gsub("n't", " not", doc)
  doc <- gsub("'ll", " will", doc)
  doc <- gsub("'re", " are", doc)
  doc <- gsub("'ve", " have", doc)
  doc <- gsub("'m", " am", doc)
  doc <- gsub("'d", " would", doc)
  # 's could be 'is' or could be possessive: it has no expansion
  doc <- gsub("'s", "", doc)
  doc <- gsub("br", "", doc)  
  return(doc)
}
```

```r
imdb$review <- sapply(imdb$review, fix.contractions)
```

```r
# function to remove special characters
removeSpecialChars <- function(x) gsub("[^a-zA-Z0-9 ]", " ", x)
# remove special characters
imdb$review <- sapply(imdb$review, removeSpecialChars)
```

```r
# convert everything to lower case
imdb$review <- sapply(imdb$review, tolower)
```

###Text Analysis

Now we can look into doing text analysis on the IMDB reviews. We can unnest the reviews into words to get a better understanding of the data. Then, we can remove stop words to ensure that no words that are used commonly in the english language are counted in the analysis.

```r
full_word_count <- imdb %>%
    unnest_tokens(word, review) %>%
    anti_join(stop_words) %>%
    group_by(word) %>%
    summarise(num_words = n()) %>%
    arrange(desc(num_words)) 
```

Now we can get an idea of how many of each word is contained in the data. Below, we can see that the words 'movie' and 'film' appear most often. We can also see that 'bad' is in the top 5 words. This may indicate that the dataset has more negative sentiment than positive.

```r
head(full_word_count)
```

```r
head(full_word_count) %>%
    ggplot(aes(x = word, y = num_words )) +
        geom_col() +
        ylab("Count") + 
        xlab("Word") +
        ggtitle("Word Count Distribution")
```

```r
imdb_words_filtered <- imdb %>%
    mutate(linenumber = row_number()) %>%
    unnest_tokens(word, review) %>%
    anti_join(stop_words)
```

Below is another way to visualize the word counts of the data. The wordcloud below is interactive. Hover your mouse above the words to see the count. Only the top 100 words are used.

```r
imdb_word_counts <- imdb_words_filtered %>%
    count(word, sort = T)

wordcloud2(imdb_word_counts[1:100,], size = .5)
```

Now lets determine the sentiment of the words used in the dataset. Note, this does not get at the sentiment of a review. This just shows the sentiment of the words used in all reviews. There are 3 lexicons for determining sentiment: afinn, bing, and nrc.

The bing lexicon uses a binary categorization model that sorts words into positive or negative.The afinn lexicon grades words between -5 and 5. The nrc lexicon categorizes sentiment words into positive, negative, anger, anticipation, disgust, fear, joy, sadness, surprise and trust.

Below is a comparison of the three lexicons.

```r
afinn <- imdb_words_filtered  %>% 
  inner_join(get_sentiments("afinn")) %>% 
  group_by(index = linenumber %/% 80) %>% 
  summarise(sentiment = sum(value)) %>% 
  mutate(method = "AFINN")

bing_and_nrc <- bind_rows(imdb_words_filtered %>% 
                            inner_join(get_sentiments("bing")) %>%
                            mutate(method = "Bing et al."),
                          imdb_words_filtered%>% 
                            inner_join(get_sentiments("nrc") %>% 
                                         filter(sentiment %in% c("positive", 
                                                                 "negative"))) %>%
                            mutate(method = "NRC")) %>%
  count(method, index = linenumber %/% 80, sentiment) %>%
  spread(sentiment, n, fill = 0) %>%
  mutate(sentiment = positive - negative)


bind_rows(afinn, 
          bing_and_nrc) %>%
  ggplot(aes(index, sentiment, fill = method)) +
  geom_col(show.legend = FALSE) +
  facet_wrap(~method, ncol = 1, scales = "free_y")
```

Let's look more in depth at the bing sentiment lexicon. The bing lexicon only labels words as negative or positive. Below we can see the count and sentiment of a given word. We see that the word 'bad' has a negative sentiment, and occurs over 14000 times in the reviews.

```r
bing_word_counts <- imdb_words_filtered %>%
  inner_join(get_sentiments("bing")) %>%
  count(word, sentiment, sort = TRUE) %>%
  ungroup()

head(bing_word_counts)
```

###Sentiment Analysis with SentimentAnalysis Package in R

Now that we have explored the data, let's take a look at Sentiment Analysis. Below, we can see that the SentimentAnalysis package scores sentiment on a scale of -1 to 1. Therefore, we must change our negative labels to -1 and the positive labels to 1. We can analyze the sentiment of the IMDB reviews by using the analyzeSentiment function. 

```r
sentiment <- analyzeSentiment(imdb$review[1:10])

sentiment$SentimentQDAP #dictionary based sentiment accroding to QDAP

convertToDirection(sentiment$SentimentQDAP)

imdb$sentiment[imdb$sentiment == 'positive'] = 1
imdb$sentiment[imdb$sentiment == 'negative'] = -1
```

After assigning the analyzed sentiment to sentiment, we can then compare it to the actual sentiment. Based of the first 10 reviews, the SentimentAnalysis package is about 70% accurate.

```r
imdb$sentiment <- as.numeric(imdb$sentiment)
compareToResponse(sentiment, imdb$sentiment[1:10])

plotSentimentResponse(sentiment$SentimentQDAP, imdb$sentiment[1:10])
```

Now, let's train a model via generate dictionary. This allows us to assign sentiment to each review via a dictionary. This will be our own sentiment dictionary. 

```r
set.seed(101) 
sample <- sample.int(n = nrow(imdb), size = floor(.75*nrow(imdb)), replace = F)
train <- imdb[sample, ]
test  <- imdb[-sample, ]

dict <- generateDictionary(train$review, train$sentiment)

summary(dict)
```

```r
compareDictionaries(dict,
                    loadDictionaryQDAP())
```

Now we can use the dictionary we created to predict the test set. Here, we can see that the model is 76% accurate.

```r
pred <- predict(dict, test$review)

compareToResponse(pred, test$sentiment)
```

```r
plotSentimentResponse(pred,test$sentiment)
```
